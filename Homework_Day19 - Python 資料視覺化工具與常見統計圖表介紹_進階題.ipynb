```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('boston.csv')
data
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
```

```python
# 利用matplotlib畫圖
plt.boxplot(data)
plt.show()

# 利用pandas畫圖
data.plot.box()
# 11(TAX)、13(B)欄的中位數在300~400之間
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
```

```python
# 利用matplotlib畫圖
plt.scatter(data.NOX,data.DIS,s=10)
plt.show()

# 利用pandas畫圖
data.plot.scatter(x='NOX',y='DIS')

# NOX與DIS呈現負相關，當NOX越大，DIS越小。
```
