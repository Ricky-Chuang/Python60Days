### 1. [簡答題] 請問下列兩種將 Array 轉換成 List 的方式有何不同？

```
print('list(a): ', list(a))
print('tolist(): ', a.tolist())
```


````python
import numpy as np
a = np.random.randint(10, size=6)
b = np.random.randint(10, size=(3,4))

print(a.tolist())
print(list(a))
print(b.tolist())
print(list(b))
````

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
````

````python
def transform(iterable):
    if type(iterable) != np.ndarray:
        return iterable    
    newlist =[]
    for x in iterable :
        newlist.append(tolist(x))
    return list(newlist)

print(transform(a))
````

````python
print(transform(b))
````

````python
print(transform(c))
````
