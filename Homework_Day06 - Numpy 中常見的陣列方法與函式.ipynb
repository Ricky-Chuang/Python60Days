### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
```



````python
import numpy as np
a = np.random.random((5, 5))
print("a :",a ,"\n\n----------------------------------\n")
print(a.sum()) 
print(np.sum(a))
print(sum(a))

# a.sum() 和 np.sum()都是屬於numpy專屬的，結果相同。
# sum(a)則是array和list都能使用，但他是將array根據axis=0(欄)來做加總。
````

### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
# 記得先 Import 正確的套件
import numpy as np
````

````python
A = np.random.random((5, 5))

A = (A-A.min()) / (A.max() - A.min())
print(A)
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
a = np.arange(1,11,1)
a[a==a.max()] = -1
print(a)

print("\n---------------------\n")

a = np.arange(1,11,1)
a[a.argmax()] = -1
print(a)
````
